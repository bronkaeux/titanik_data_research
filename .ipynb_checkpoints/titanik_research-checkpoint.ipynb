```python
penalty = ['l1', 'l2', 'elasticnet']
l1_ratio = np.linspace(0,1,20)
C = np.logspace(0,10,20)

param_grid = {'penalty': penalty,
             'l1_ratio': l1_ratio,
             'C': C}
```

```python
grid_model = GridSearchCV(log_model,param_grid=param_grid)
```

```python
grid_model.fit(scaled_X_train,y_train)
```
